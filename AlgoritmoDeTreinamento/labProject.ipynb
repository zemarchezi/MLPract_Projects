Neste projeto devemos seguir o exemplo do algoritmo de binarização apresentado em aula, realizar a implementação em Python para transformar uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco.


### Importa os módulos necessários

```python
from PIL import Image
```

```python

# Converter para escala de cinza
def converte_para_cinza(imagem):
    largura, altura = imagem.size
    imagem_cinza = Image.new("L", (largura, altura))  # Criar nova imagem em escala de cinza

    for y in range(altura):
        for x in range(largura):
            # Obtem os valores RGB do pixel
            r, g, b = imagem.getpixel((x, y))
            # Aplica a fórmula para conversão para tons de cinza
            valor_cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
            # Atribui o valor calculado ao pixel da imagem cinza
            imagem_cinza.putpixel((x, y), valor_cinza)

    return imagem_cinza
```

```python
def binarizar_imagem(imagem_cinza, limiar=127):
    largura, altura = imagem_cinza.size
    imagem_binaria = Image.new("1", (largura, altura))  # Cria nova imagem binária (1 bit por pixel)

    for y in range(altura):
        for x in range(largura):
            # Obteem o valor de cinza do pixel
            valor_cinza = imagem_cinza.getpixel((x, y))
            # Verifica se está acima ou abaixo do limiar
            valor_binario = 1 if valor_cinza > limiar else 0  # 1 para branco, 0 para preto
            # Atribui o valor binário ao pixel da imagem binária
            imagem_binaria.putpixel((x, y), valor_binario)

    return imagem_binaria
```

```python
path_imagem= "./"
nome_imagem = "bicicleta.jpg"
imagem_original = Image.open(f"{path_imagem}{nome_imagem}")
```

### Converte para cinza

```python
imagem_cinza = converte_para_cinza(imagem_original)
```

### Converte para binário

```python
imagem_binaria = binarizar_imagem(imagem_cinza)
```

```python
imagem_cinza.save("imagem_cinza.png")
imagem_binaria.save("imagem_binaria.png")
```

```python
imagem_original.show(title="Imagem Original")
imagem_cinza.show(title="Imagem em Tons de Cinza")
imagem_binaria.show(title="Imagem Binarizada")
```
